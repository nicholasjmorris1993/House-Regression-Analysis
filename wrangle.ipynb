```python
import pandas as pd
```

```python
# get the data
df = pd.read_csv("prices.csv")
```

```python
# check out the data
df.head()
```

```python
# remove the id column
df = df.drop(columns="Id")
```

```python
# replace missing values with None for string columns
for col in df.columns:
    if df[col].dtypes == "object" and df[col].isna().any():
        df[col] = df[col].fillna("None")
```

```python
# convert YrSold to a date
df["DateSold"] = df["YrSold"].astype(str) + "-01-01"
df = df.drop(columns="YrSold")
```

```python
# fill in missing values for LotFrontage, MasVnrArea, and GarageYrBlt with 0
df["LotFrontage"] = df["LotFrontage"].fillna(0)
df["MasVnrArea"] = df["MasVnrArea"].fillna(0)
df["GarageYrBlt"] = df["GarageYrBlt"].fillna(0)
```

```python
# shuffle the data
df = df.sample(frac=1, random_state=0).reset_index(drop=True)
```

```python
# check out the data
df.head()
```

```python
# export the data
df.to_csv("house.csv", index=False)
```
